```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Candy Crush Saga.csv')
```

```python
from transformers import pipeline
from transformers import DistilBertTokenizer , DistilBertForSequenceClassification
tokeniser = DistilBertTokenizer.from_pretrained('distilbert/distilbert-base-uncased-finetuned-sst-2-english')
model = DistilBertForSequenceClassification.from_pretrained('distilbert/distilbert-base-uncased-finetuned-sst-2-english')
```

```python

nlp = pipeline('sentiment-analysis',model=model,tokenizer=tokeniser)
```

```python
texts = list(df.content.values)
results = nlp(texts)
```

```python
for text, result, score in zip(texts, results, df.score.values):
  print("Text: ", text)
  print("Result: ", result)
  print("Score: ", score)
```

```python
df['sentiment'] = [result['label'] for result in results]
df
```
